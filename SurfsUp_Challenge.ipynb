```python
# Dependencies
import numpy as np
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 

june = session.query(Measurement.tobs).filter(func.strftime('%m', Measurement.date) == "06").all()
```

```python
#  3. Convert the June temperatures to a list.
june_temps = list(np.ravel(june))

june_temps
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_temps_df = pd.DataFrame(june_temps,columns= ['Temperature'])
june_temps_df
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_temps_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.

december = session.query(Measurement.tobs).filter(func.strftime('%m', Measurement.date) == "12").all()
```

```python
# 7. Convert the December temperatures to a list.
dec_temps = list(np.ravel(december))

dec_temps
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
dec_temps_df = pd.DataFrame(dec_temps,columns= ['Temperature'])
dec_temps_df
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
dec_temps_df.describe()
```

# Additional Queries: Drop the data that has too high elevation

```python
#look at elevation
results = session.query(Station.station, Station.elevation)

for row in results:
    print(row)
```

```python
#drop the highest elevations
low_elevation = session.query(Station.station, Station.elevation).filter(Station.station != 'USC00518838').filter(Station.station != 'USC00516128').all()

for row in low_elevation:
    print(row)
```

```python
#create an elevation DataFrame
elevation_df = pd.read_sql_query(sql = session.query(Station.station, Station.elevation).filter(Station.station != 'USC00518838').filter(Station.station != 'USC00516128').statement, con = engine)
elevation_df
```

```python
#describe the new june data

june_elevation = session.query(Measurement.station, Measurement.tobs).\
    filter(func.strftime('%m', Measurement.date) == "06").\
    filter(Measurement.station != 'USC00518838').\
    filter(Measurement.station != 'USC00516128').all()

june_elevation_df = pd.DataFrame(june_elevation,columns= ['Station', 'Temperature'])

print(june_elevation_df)

june_elevation_df.describe()
```

```python
#describe the new December data
dec_elevation = session.query(Measurement.station, Measurement.tobs).\
    filter(func.strftime('%m', Measurement.date) == "12").\
    filter(Measurement.station != 'USC00518838').\
    filter(Measurement.station != 'USC00516128').all()


dec_elevation_df = pd.DataFrame(dec_elevation, columns = ['Station', 'Temperature'])

print(dec_elevation_df)

dec_elevation_df.describe()
```

```python
#Look at precipitation taking out high elevation for June

june_elv_prcp = session.query(Measurement.station, Measurement.prcp).\
    filter(func.strftime('%m', Measurement.date) == "06").\
    filter(Measurement.station != 'USC00518838').\
    filter(Measurement.station != 'USC00516128').all()

june_elv_prcp_df = pd.DataFrame(june_elv_prcp,columns= ['Station', 'Precipitation'])

print(june_elv_prcp_df)

june_elv_prcp_df.describe()
```
